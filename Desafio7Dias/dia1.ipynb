```python
#Criando nova coluna baseada nas informações de localização dos livros para categorizar eles(CDU)
def get_cdu_class(location):
    if 0 <= location <= 99:
        return "Generalidades. Ciência e conhecimento"
    elif 100 <= location <= 199:
        return "Filosofia e psicologia"
    elif 200 <= location <= 299:
        return "Religião"
    elif 300 <= location <= 399:
        return "Ciências sociais"
    elif 400 <= location <= 499:
        return "Classe vaga. Provisoriamente não ocupada"
    elif 500 <= location <= 599:
        return "Matemática e ciências naturais"
    elif 600 <= location <= 699:
        return "Ciências aplicadas"
    elif 700 <= location <= 799:
        return "Belas artes"
    elif 800 <= location <= 899:
        return "Linguagem. Língua. Linguística"
    elif 900 <= location <= 999:
        return "Geografia. Biografia. História"
    else:
        return "Código de localização inválido"

# Aplicar a função para criar a nova coluna
emprestimos_completos['classe_geral_CDU'] = emprestimos_completos['localizacao'].apply(get_cdu_class)
emprestimos_completos
```

```python
#Função boxplot
def gera_box_plot(dataset, x, y, titulo, subtitulo):
    '''
    Esta função irá gerar um gráfico de boxplot.

    Dataset = conjunto de dados do gráfico
    x = valor do eixo x do gráfico
    y = valor do eixo y do gráfico
    titulo = título do gráfico
    subtitulo = subtítulo do texto
    '''

    sns.set_theme(style="darkgrid", palette="Blues", font_scale=1.3)
    plt.figure(figsize=(16, 10))

    ax = sns.boxplot(y=y, x=x, data=dataset, color='#4171EF')
    ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',', '.')))

    plt.ylim(0, max(dataset[y]) * 1.1)  # Definir o limite do eixo y
    plt.xlabel(None)
    plt.ylabel(None)

    ax.set_title(titulo + "\n", size=20, loc='left', weight='bold')
    ax.text(s=subtitulo, x=-0.5, y=max(dataset[y]) * 1.11, fontsize=18, ha='left', color='gray')
```

```python
#Tabela
gera_box_plot(alunos_graduacao_acervo_circulante,"ano","quantidade","Distribuição de emprestimos mensais","Por alunos de graduação no acervo circulante")
```
